# Importing Required Libraries

```python
# Importing the NumPy library for numerical operations
import numpy as np

# Importing the Pandas library for data manipulation and analysis
import pandas as pd

# Importing the Matplotlib library for data visualization
import matplotlib.pyplot as plt

# Importing the Seaborn library for statistical data visualization
import seaborn as sns
```

# Loading the Dataset

```python
# Reading the dataset from the 'housing.csv' file into a Pandas DataFrame
data = pd.read_csv('housing.csv')
```

```python
data
```

# Data Preprocessing

```python
# Displaying a concise summary of the DataFrame, including the data types and non-null values of each column
data.info()
```

```python
# Generating descriptive statistics of the DataFrame, including count, mean, standard deviation, min, and max values for each numerical column
data.describe()
```

```python
# Calculating the total number of missing values in each column of the DataFrame
data.isnull().sum()
```

```python
# Dropping all rows with missing values from the DataFrame and modifying it in place
data.dropna(inplace=True)
```

```python
# Displaying a concise summary of the DataFrame after dropping rows with missing values
data.info()
```

# Splitting the Dataset into Training and Testing Sets

```python
# Importing the train_test_split function from scikit-learn for splitting the data into training and testing sets
from sklearn.model_selection import train_test_split

# Separating the features (all columns except 'median_house_value') into X
X = data.drop(['median_house_value'], axis=1)

# Separating the target variable (the 'median_house_value' column) into y
y = data['median_house_value']
```

```python
# Displaying the features DataFrame (X) which contains all columns except 'median_house_value'
X
```

```python
# Displaying the target variable Series (y) which contains only the 'median_house_value' column
y
```

```python
# Splitting the data into training and testing sets with 20% of the data reserved for testing
# X_train and y_train are used for training the model, while X_test and y_test are used for evaluating the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Exploratory Data Analysis (EDA)

```python
# Combining the training features (X_train) and the training target variable (y_train) into a single DataFrame
train_data = X_train.join(y_train)
```

```python
# Displaying the combined training DataFrame which includes both features and target variable
train_data
```

```python
# Plotting histograms for each feature and target variable in the training DataFrame to visualize their distributions
train_data.hist(figsize=(15, 8))
```

```python
# Applying a logarithmic transformation to the specified features to handle skewness and normalize their distribution
# Adding 1 before taking the log to avoid issues with zero values

train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
train_data['median_income'] = np.log(train_data['median_income'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# Counting the number of occurrences of each category in the 'ocean_proximity' column of the training DataFrame
train_data.ocean_proximity.value_counts()
```

```python
# One-hot encoding the 'ocean_proximity' categorical column and joining the resulting dummy variables with the training DataFrame
# The original 'ocean_proximity' column is then dropped
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop('ocean_proximity', axis=1)
```

```python
train_data
```

```python
# Computing the correlation matrix of the training DataFrame to understand the relationships between different features and the target variable
train_data.corr()
```

```python
# Creating a heatmap to visualize the correlation matrix of the training DataFrame
# The heatmap will include annotations to show the correlation coefficients
plt.figure(figsize=(12, 8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
# Creating a scatter plot to visualize the relationship between latitude and longitude in the training data
# The color of the points represents the median house value, using a cool-to-warm color palette
plt.figure(figsize=(12, 8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='coolwarm')
plt.title('Scatter Plot of Latitude vs. Longitude')
plt.show()
```

# Feature Engineering

```python
# Creating new feature columns for the training DataFrame:
# 'bedroom_ratio' represents the ratio of total bedrooms to total rooms, indicating bedroom density
# 'house_ratio' represents the ratio of total rooms to households, indicating the average number of rooms per household
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['house_ratio'] = train_data['total_rooms'] / train_data['households']
```

```python
# Displaying the updated training DataFrame which now includes the newly created feature columns 'bedroom_ratio' and 'house_ratio'
train_data
```

```python
# Creating a heatmap to visualize the updated correlation matrix of the training DataFrame
# The heatmap includes annotations to display the correlation coefficients, showing how the new features relate to others
plt.figure(figsize=(12, 8))
sns.heatmap(train_data.corr(), annot=True)
```

# Training the Model

```python
# Importing the LinearRegression model and metrics for evaluating model performance
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Preparing the training data by separating features (X_train) and target variable (y_train)
X_train = train_data.drop(['median_house_value'], axis=1)
y_train = train_data['median_house_value']
```

```python
# Initializing and training a Linear Regression model using the training features (X_train) and target variable (y_train)
reg = LinearRegression()
reg.fit(X_train, y_train)
```

# Evaluating the Model

```python
# Combining the testing features (X_test) and the testing target variable (y_test) into a single DataFrame
test_data = X_test.join(y_test)

# Applying the same logarithmic transformation to the test data as done for the training data
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)
test_data['median_income'] = np.log(test_data['median_income'] + 1)

# One-hot encoding the 'ocean_proximity' categorical column and joining the resulting dummy variables with the test data
# The original 'ocean_proximity' column is then dropped
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity).astype(int)).drop('ocean_proximity', axis=1)

# Creating new feature columns for the test data:
# 'bedroom_ratio' represents the ratio of total bedrooms to total rooms
# 'house_ratio' represents the ratio of total rooms to households
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['house_ratio'] = test_data['total_rooms'] / test_data['households']
```

```python
# Preparing the test data by separating features (X_test) and target variable (y_test)
X_test = test_data.drop(['median_house_value'], axis=1)
y_test = test_data['median_house_value']
```

```python
# Evaluating the performance of the trained Linear Regression model on the test data
# The score method returns the coefficient of determination (R^2) of the prediction
reg.score(X_test, y_test)
```

```python
# Predicting the target values using the trained model
y_pred = reg.predict(X_test)

# Creating a DataFrame for the predictions to include true and predicted values
plot_data = pd.DataFrame({'True Values': y_test, 'Predictions': y_pred})

# Scatter plot of true values vs. predicted values
plt.figure(figsize=(12, 8))
plt.scatter(plot_data['True Values'], plot_data['Predictions'], alpha=0.5)
plt.plot([plot_data['True Values'].min(), plot_data['True Values'].max()],
         [plot_data['True Values'].min(), plot_data['True Values'].max()],
         color='red', linestyle='--')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs. Predicted Values')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculating Mean Squared Error (MSE) and Mean Absolute Error (MAE) for the test data
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```

```python
from sklearn.metrics import r2_score

# Calculating RÂ² score for the test data
r2 = r2_score(y_test, y_pred)

print(f"RÂ² Score: {r2}")
```
